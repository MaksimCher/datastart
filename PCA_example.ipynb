```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_dim = 2
n_elem = 10
data = np.zeros((n_elem, n_dim))
```

```python
data[:, 0] = np.arange(n_elem)
data[:, 1] = np.arange(n_elem)
```

```python
data = data + np.random.randn(10, 2) * 2
```

```python
projections = np.zeros((10, 2))
projections[:, 0] = data.mean(axis=1)
projections[:, 1] = data.mean(axis=1)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.gca()

ax.scatter(data[:, 0], data[:, 1], s=100)
ax.scatter(projections[:, 0], projections[:, 1], s=100, c='r')
ax.plot(np.arange(-3, 12), np.arange(-3, 12))

for index in range(n_elem):
    ax.plot([data[index, 0], projections[index,0]], [data[index, 1], projections[index,1]], 'r--')

ax.tick_params(labelbottom='off', labelleft='off')    
    
```

```python
fig.savefig('pca_example.png')
```
