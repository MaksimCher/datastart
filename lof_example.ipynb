```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
%matplotlib inline
from sklearn.neighbors import LocalOutlierFactor
```

```python
data = np.random.randn(100,2)
```

```python
model = LocalOutlierFactor().fit(data)
all_predictions = model._decision_function(data)
min_score_index = all_predictions.argmin()
max_score_index = all_predictions.argmax()
marker_size = 100
score_to_circle_multiplier = 20000
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.gca()

ax.scatter(data[:, 0], data[:, 1], s=marker_size)
ax.scatter(data[min_score_index, 0], data[min_score_index, 1], 
            s=-all_predictions[min_score_index] * score_to_circle_multiplier, 
            facecolors='none', edgecolors='r')
ax.scatter(data[min_score_index, 0], data[min_score_index, 1], s=marker_size, c='r')
ax.tick_params(labelbottom='off', labelleft='off')    
fig.savefig('lof_1.png')
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.gca()

ax.scatter(data[:, 0], data[:, 1], s=marker_size)
ax.scatter(data[max_score_index, 0], data[max_score_index, 1], 
            s=-all_predictions[max_score_index] * score_to_circle_multiplier, 
            facecolors='none', edgecolors='r')

ax.scatter(data[max_score_index, 0], data[max_score_index, 1], 
            s=marker_size, c='r')

ax.tick_params(labelbottom='off', labelleft='off')    
fig.savefig('lof_2.png')
```
